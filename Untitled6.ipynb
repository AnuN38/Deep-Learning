```python
import cv2 as cv
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images.jpg')
```

```python
img = cv.imshow('dog',img)
```

```python
imgplot = plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)
```

```python
imgplot = plt.imshow(gray)
```

```python
blur = cv.GaussianBlur(img, (7,7), cv.BORDER_DEFAULT)
#cv.imshow('Blur', blur)
```

```python
imgplot = plt.imshow(blur)
```

```python
canny = cv.Canny(blur,100,150)
```

```python
imgplt = plt.imshow(canny)
```

```python
dilated = cv.dilate(canny,(7,7),iterations=3)
```

```python
imgplt = plt.imshow(dilated)
```

```python
eroded = cv.erode(dilated, (7,7), iterations=3)
```

```python
imgplt = plt.imshow(eroded)
```
