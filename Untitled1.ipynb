```python
import pandas as pd
```

```python
df = pd.read_csv('CarPrice_Assignment (2).csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['CarName'] = le.fit_transform(df['CarName'])
```

```python
df['fueltype'] = le.fit_transform(df['fueltype'])
```

```python
df['aspiration'] = le.fit_transform(df['aspiration'])
```

```python
df['doornumber'] = le.fit_transform(df['doornumber'])
```

```python
df['carbody'] = le.fit_transform(df['carbody'])
```

```python
df['drivewheel'] = le.fit_transform(df['drivewheel'])
```

```python
df['enginelocation'] = le.fit_transform(df['enginelocation'])
```

```python
df['fuelsystem'] = le.fit_transform(df['fuelsystem'])
```

```python
df['enginetype'] = le.fit_transform(df['enginetype'])
```

```python
df['cylindernumber'] = le.fit_transform(df['cylindernumber'])
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_data = scalar.fit_transform(df)
scaled_data = pd.DataFrame(scaled_data,columns=df.columns)
scaled_data.describe()
```

```python
#Doing pca
from sklearn.decomposition import PCA
pca = PCA(n_components = 10)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python
x_pca
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c ="red")
 
plt.scatter(x_pca[:,2],x_pca[:,3], c ="yellow")
plt.scatter(x_pca[:,4],x_pca[:,5], c="blue")
plt.scatter(x_pca[:,6],x_pca[:,7], c="green")
plt.scatter(x_pca[:,8],x_pca[:,9], c="pink")
plt.show()
```
