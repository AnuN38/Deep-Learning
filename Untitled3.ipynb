```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,Conv2D,MaxPool2D
from tensorflow.keras import layers
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
def precision(y_true,y_pred):
    true_positive=k.sum(k.round(k.clip(y_true*y_pred,0,1)))
    predicted_positives=k.sum(k.round(k.clip(y_pred,0,1)))
    precision=true_positive/(predicted_positives+k.epsilon())
    return precision

def recall(y_true,y_pred):
    true_positive=k.sum(k.round(k.clip(y_true*y_pred,0,1)))
    possible_positive=k.sum(k.round(k.clip(y_true,0,1)))
    recall=true_positive/(possible_positive+k.epsilon())
    return recall
```

```python
import keras.backend as k
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy',precision,recall])
```

```python
history=model.fit(X_train,y_train,epochs=5,verbose=1,validation_data=(X_test,y_test))
```

```python
model.evaluate(X_test,y_test)
```
