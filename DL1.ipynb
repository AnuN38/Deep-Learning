```python
import pandas as pd
```

```python
df = pd.read_excel('iris.xls')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['SL'].fillna(df['SL'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['SW'].fillna(df['SW'].mean(),inplace=True)
```

```python
df['PL'].fillna(df['PL'].mean(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Classification'] = le.fit_transform(df['Classification'])
```

```python
df
```

```python
import tensorflow
```

```python
a = df.to_numpy()
```

```python
a
```

```python
x = a[:,0:4]
```

```python
x
```

```python
y = a[:,4]
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X= sc.fit_transform(x)
```

```python
X
```

```python
Y = tensorflow.keras.utils.to_categorical(y,3)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from tensorflow import keras
```

```python
model = tensorflow.keras.models.Sequential()
```

```python
#Adding hidden layers
#First input layer
model.add(keras.layers.Dense(150, #number of nodes
                             input_shape=(4,),#number of input variables
                            activation ='relu'))#activation function
```

```python
#Adding second hidden layer
model.add(keras.layers.Dense(130,activation ='relu'))
```

```python
#Adding output layer
model.add(keras.layers.Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
#compile model with loss and metrics
model.compile(loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
import matplotlib.pyplot as plt
```

```python
#Trainining model
VERBOSE=1

#Set batch size
BATCH_SIZE=16
#Set number of epochs
EPOCHS=10

#Setting validation split
VALIDATION_SPLIT=0.2

#Fit model
history=model.fit(x_train,y_train,batch_size=BATCH_SIZE,epochs=EPOCHS,verbose=VERBOSE,validation_split=VALIDATION_SPLIT)
pd.DataFrame(history.history)['accuracy'].plot(figsize=(8,5))
plt.title("Accuracy improvements with epochs")
plt.show()
model.evaluate(x_test,y_test)
```

```python
import numpy as np
```

### Prediction 

```python
#Prediction
prediction_input=[[2.3,2.2,1.4,2.3]]
#scaling input

x1=sc.fit_transform(prediction_input)
pred=model.predict(x1)
#print prediction
predict=np.argmax(pred)
print('prediction is:',le.inverse_transform([predict]))
```
